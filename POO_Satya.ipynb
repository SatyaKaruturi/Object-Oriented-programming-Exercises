```
#Crée le boucle pour automatisé le choix de methode aprés au fin de pluseir tour affiche est que tu as Gagné au Appris
cnt = 0 #pour counte combien fois il fait le choix avant de gagné ou apris
dataA1 = dataAnalyst("John","Wick",33,"M","Masters")

print("Initial values: Motivation: {}, Progression: {}".format(dataA1.motivation, dataA1.progression))

while(dataA1.motivation >= 0 and dataA1.progression <= 100): 
  choosenaction = np.random.randint(0,3)
  if choosenaction == 0:
    methode = "suivreFormation"
    dataA1.suivreFormation()
  if choosenaction == 1:
    methode = "bosserPlus"
    dataA1.bosserPlus()
  if choosenaction == 2:
    methode = "echouer"
    dataA1.echouer()
  if choosenaction == 3:
    methode = "reussir"
    dataA1.reussir()
    
  cnt += 1
  
  print("Tour no",cnt)
  print("La méthode {} a été utilisée.". format(methode))
  print("motivation vaut maintenant {}" .format(dataA1.motivation))
  print("progression vaut maintenant {}" .format(dataA1.progression))
if dataA1.motivation <=0: print("BRAVO TU AS GAGNÉ !!!")
if dataA1.progression >=100: print("BRAVO TU AS APPRIS !!!")
```
